Problem Statement: As a macroeconomic analyst at the Organization for Economic Cooperation and Development (OECD), your job is to collect relevant data for analysis. It looks like you have three countries in thenorth_americadata frame and one country in thesouth_americadata frame. As these are in two separate plots, it's hard to compare the average labor hours between North America and South America. If all the countries were into the same data frame, it would be much easier to do this comparison.

Objective:Demonstrate concatenation.

```python
import pandas as pd
```

```python
nm = pd.read_csv('north_america.csv')
nm
```

```python
sm = pd.read_csv('south_america.csv')
sm
```

```python
# Combine both the dataframes, so they are easier to compare. Axis = 0 indicates row-wise concatenation
americas = pd.concat([nm,sm],axis=0)
```

```python
americas
```

```python
americas_transpose = americas.transpose()
americas_transpose
```

```python
# Plot the Labour hours
americas_transpose.columns = americas_transpose.iloc[0, :]
americas_transpose = americas_transpose.iloc[1:]
americas_transpose.plot()
```

```python
# Conclusion: Above plot shows the labour hours of Chile and Mexico are much higher than Canada and USA. 
```
